```python
import pandas as pd
from scipy.stats import zscore

# Load the CSV file
data = pd.read_csv("/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/Skinly_Thailand/Akshay_projects2/updated_thai_chunkid.csv")

# Choose a specific user ID
user_id = 3154

# Filter data for the chosen user ID
user_data = data[data['user_id'] == user_id]

# Initialize a list to store outlier data points
outliers_to_remove = []

# Iterate through each feature
for feature in data.columns[3:]:
    # Calculate z-scores for the feature
    z_scores = zscore(user_data[feature])
    
    # Identify outliers using z-score method (e.g., z-score > 3 or z-score < -3)
    outlier_indices = abs(z_scores) > 3
    
    # Get the dates of outlier data points
    outlier_dates = user_data.loc[outlier_indices, 'date']
    
    # If all outlier data points have the same date, mark them for removal
    if outlier_dates.nunique() == 1:
        outliers_to_remove.extend(outlier_dates)

# Remove outliers with the same date for all features
cleaned_data = data[~((data['user_id'] == user_id) & (data['date'].isin(outliers_to_remove)))]

# Print cleaned data
print(cleaned_data)
```

```python
import pandas as pd
from scipy.stats import zscore

# Load the CSV file
data = pd.read_csv("/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/Skinly_Thailand/Akshay_projects2/updated_thai_chunkid.csv")

# Initialize a dictionary to store outlier data points along with their dates for all users
outliers_with_dates = {}

# Iterate through each user ID
for user_id in data['user_id'].unique():
    # Filter data for the current user ID
    user_data = data[data['user_id'] == user_id]

    # Initialize a dictionary to store outlier data points along with their dates for the current user
    user_outliers_with_dates = {}

    # Iterate through each feature
    for feature in data.columns[3:]:
        # Calculate z-scores for the feature
        z_scores = zscore(user_data[feature])
        
        # Identify outliers using z-score method (e.g., z-score > 4 or z-score < -4)
        outlier_indices = abs(z_scores) > 1
        
        # Get the dates of outlier data points
        outlier_dates = user_data.loc[outlier_indices, 'date']
        
        # Store outliers along with their dates
        user_outliers_with_dates[feature] = outlier_dates.tolist()
    
    # Store outlier data for the current user in the main dictionary
    outliers_with_dates[user_id] = user_outliers_with_dates

# Print outliers with dates for each user
for user_id, user_outliers in outliers_with_dates.items():
    print("Outliers for user ID:", user_id)
    for feature, dates in user_outliers.items():
        if dates:
            print(f"Feature: {feature}, Outliers Dates: {dates}")

# Remove outliers from the data
cleaned_data = data.copy()  # Make a copy of the original data
for user_id, user_outliers in outliers_with_dates.items():
    for feature, dates in user_outliers.items():
        cleaned_data = cleaned_data[~((cleaned_data['user_id'] == user_id) & (cleaned_data['date'].isin(dates)))]

# Save the cleaned data to a new CSV file
cleaned_data.to_csv("/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/Skinly_Thailand/Akshay_projects2/outlier_and_normalisation/cleaned_data.csv", index=False)

print("Cleaned data saved to 'cleaned_data.csv'")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the first dataset
df1 = pd.read_csv('updated_thai_chunkid.csv')
# Load the second dataset
df2 = pd.read_csv('cleaned_data.csv')
df1 = df1.apply(pd.to_numeric, errors='coerce')
df2 = df2.apply(pd.to_numeric, errors='coerce')

# Determine the number of features
num_features = len(df1.columns)

# Set the number of subplots per row
subplots_per_row = 2

# Calculate the number of rows needed
num_rows = (num_features + 1) // subplots_per_row

# Create subplots
fig, axes = plt.subplots(num_rows, subplots_per_row, figsize=(15, 5*num_rows))

# Flatten axes if needed
if num_rows == 1:
    axes = [axes]

# Define colors for original and cleaned data
color_original = 'blue'
color_cleaned = 'green'

# Plot histograms for each feature
for idx, column in enumerate(df1.columns):
    if column not in ['user_id', 'date', 'chunkid']:
        row_idx = idx // subplots_per_row
        col_idx = idx % subplots_per_row
        
        # Plot histograms for the current feature from both datasets
        axes[row_idx][0].hist(df1[column].dropna(), bins=20, alpha=0.5, color=color_original, label='Original Data')
        axes[row_idx][1].hist(df2[column].dropna(), bins=20, alpha=0.5, color=color_cleaned, label='Cleaned Data')
        
        # Set title and labels
        axes[row_idx][0].set_title(column + ' (Original Data)')
        axes[row_idx][1].set_title(column + ' (Cleaned Data)')
        axes[row_idx][0].set_xlabel('Value')
        axes[row_idx][0].set_ylabel('Frequency')
        axes[row_idx][1].set_xlabel('Value')
        axes[row_idx][1].set_ylabel('Frequency')
        axes[row_idx][0].legend()
        axes[row_idx][1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
