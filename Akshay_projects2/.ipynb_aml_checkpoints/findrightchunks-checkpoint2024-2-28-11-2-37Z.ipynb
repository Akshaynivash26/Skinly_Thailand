```python
# Ensure all_chunks is sorted by 'chunk_start_date' for efficient searching
all_chunks = all_chunks.sort_values(by='chunk_start_date')
 
# Create a new empty column named 'chunkiduser'
dateuser_df['chunkiduser'] = ''
 
# Define a function to get chunk ID for each row
def get_chunk_id(row):
    user_id = row['user_id']
    date = row['date']
    # Filter rows from all_chunks DataFrame
    chunk = all_chunks[(all_chunks['user_id'] == user_id) & 
                       (all_chunks['chunk_start_date'] <= date) & 
                       (all_chunks['chunk_end_date'] >= date)]
    # Return chunk_id if found, otherwise return None
    if not chunk.empty:
        return chunk.iloc[0]['chunk_id']
    else:
        return None
 
# Apply the function to each row and assign chunk ID to the new column
dateuser_df['chunkiduser'] = dateuser_df.apply(get_chunk_id, axis=1)
```

```python
# Ensure all_chunks is sorted by 'chunk_start_date' for efficient searching
all_chunks = all_chunks.sort_values(by='chunk_start_date')
 
# Define a function to get chunk ID for each row
def get_chunk_id(row):
    user_id = row['user_id']
    date = row['date']
    # Filter rows from all_chunks DataFrame
    chunk = all_chunks[(all_chunks['user_id'] == user_id) & 
                       (all_chunks['chunk_start_date'] <= date) & 
                       (all_chunks['chunk_end_date'] >= date)]
    # Return chunk_id if found, otherwise return None
    if not chunk.empty:
        return chunk.iloc[0]['chunk_id']
    else:
        return None
 
# Apply the function to each row and assign chunk ID to a new column
skin_thai['chunkiduser'] = skin_thai.apply(get_chunk_id, axis=1)
 
# Reorder columns to place 'chunkiduser' as the fourth column
cols = skin_thai.columns.tolist()
cols.insert(2, cols.pop(cols.index('chunkiduser')))
skin_thai = skin_thai.reindex(columns=cols)

```
