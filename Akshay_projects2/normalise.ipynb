```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv('/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/Skinly_Thailand/Akshay_projects2/updated_thai_chunkid.csv')

```

```python
# Define a function to remove outliers using the z-score method
def remove_outliers_zscore(df, features, threshold=3):
    z_scores = np.abs((df[features] - df[features].mean()) / df[features].std())
    df = df[(z_scores < threshold).all(axis=1)]
    return df
```

```python

# Define a list of features to check for outliers
features = ['moisture_cheek', 'moisture_forearm', 'moisture_forehead', 'shininess_cheek', 'shininess_forehead', 'evenness_cheek', 'evenness_forearm', 'evenness_red_cheek', 'evenness_red_forearm', 'evenness_red_forehead', 'ita_cheek', 'ita_forearm', 'ita_forehead', 'melanin_cheek', 'melanin_forearm', 'melanin_forehead', 'vessels_den_cheek', 'vessels_den_forearm', 'vessels_den_forehead', 'skin_texture_cheek', 'skin_texture_forearm', 'skin_texture_forehead', 'fine_lines_cheek', 'fine_lines_forearm', 'fine_lines_forehead', 'deep_lines_forehead', 'cheek_age_dif', 'forearm_age_dif', 'forehead_age_dif', 'G_porph_L_MEAN_cheek', 'G_porph_L_MEAN_forehead', 'G_porph_NUMBER_cheek', 'G_porph_NUMBER_forehead', 'R_porph_L_MEAN_cheek', 'R_porph_L_MEAN_forehead', 'R_porph_NUMBER_cheek', 'R_porph_NUMBER_forehead', 'Y_porph_L_MEAN_cheek', 'Y_porph_L_MEAN_forehead', 'Y_porph_NUMBER_cheek', 'Y_porph_NUMBER_forehead', 'redness_cheek', 'redness_forearm', 'redness_forehead']

```

```python
# Remove outliers using the z-score method
df = remove_outliers_zscore(df, features)
```

```python

# Save the updated data to a new CSV file
df.to_csv('outliers_removed_thai.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
dfout= pd.read_csv('/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/Skinly_Thailand/Akshay_projects2/outliers_removed_thai.csv')

# Calculate the z-scores for each column
z_scores = (dfout[dfout.columns[3:]] - dfout[dfout.columns[3:]].mean()) / dfout[dfout.columns[3:]].std()

# Scale the data between 0 and 100
scaled_data = z_scores * 50 + 50
# Clip the scaled values to a maximum of 100
scaled_data = scaled_data.clip(upper=100)

# Replace the original data with the scaled data
dfout[dfout.columns[3:]] = scaled_data

# Save the standardized and scaled data to a new CSV file
dfout.to_csv('z_score_normalized.csv', index=False)
```
